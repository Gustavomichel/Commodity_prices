```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"data\commodity_prices.csv").dropna(axis="columns")
df
```

```python
df_mean = df.mean().drop("Year")
df_mean_100 = df_mean[df_mean>100]
df_mean_50 = df_mean[df_mean<=50]
df_mean_50
```

```python
sns.set(style="darkgrid", palette="dark")
df_mean_100.plot(kind="bar", 
            ylabel='Historic Prices', xlabel='Products', 
            title='Products historical price mean (1960-2021) > 100U$', fontsize=9
            , figsize=(8, 5), color='green')
```

```python
sns.set(style="darkgrid", palette="dark")
df_mean_50.plot(kind="bar", 
            ylabel='Historic Prices', xlabel='Products', 
            title='Products historical price mean (1960-2021) < 50U$', fontsize=9
            , figsize=(8, 5), color='green')
```

```python
df_dif = df.loc[61] - df.loc[0]
df_dif = pd.DataFrame(df_dif).drop("Year")

df_md = df.median().drop("Year")
```

```python
sns.set(style="darkgrid", palette="dark")

df_dif.plot(kind="bar", figsize=(8, 6), title='Historical price difference', ylabel='difference', xlabel='Products', color='red', legend=False)
```

```python
sns.set(style="darkgrid", palette="dark")

df_md.plot(kind="bar", figsize=(8, 6), title='median prices', ylabel='difference', xlabel='Products', color='red', legend=False)
```
